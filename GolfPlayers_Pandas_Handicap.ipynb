```python
class Handicap():
    #docstring for handicap
    def __init__(self,scores):
        self.scores = scores
        self.handicap = 0

    def h_total(self):
        # initializing K  
        K = 5
        # Inverse K slice Sum 
        # using list slicing + sum() 
        res = sum(self.scores[-K:]) 
        #round the handicap
        self.handicap = float((res/5)-36)
        return self.handicap
```

```python
def handicap_calc(scores):
        # initializing K  
        K = 5
        # Inverse K slice Sum 
        # using list slicing + sum() 
        res = sum(scores[-K:]) 
        #handicap as a float, round before you write row
        return float((res/5)-36)
```
